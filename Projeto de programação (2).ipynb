```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Livro-1-_1_.csv')
df.head()
```

```python
#extraindo as colunas que não são importantes para o estudo em causa
#colunas excluidas: %arremessos, %arremessos de 3 pontos, %Lances livres
df = df.drop(columns=['%AC', '3P%', '%LL' ])
```

```python
df.head()
```

```python
#renomeando as colunas para tornar mais fácil o estudo desta
df.columns = ['Posição', 'Jogador', 'Equipa', 'Partidas jogadas', 'Minutos por jogo', 'Pontos por jogo', 'Arremessos convertidos', 'Tentativa de arremessos', 'Arremesso de 3 pontos convertidos', 'Tentativa de arremessos de 3 pontos', 'Lances livre convertido', 'Tentativa de arremessos livres']
df.head()
```

```python
#%matplotlib notebook
%matplotlib inline

tentativa_arremessos = df ['Tentativa de arremessos'].tolist()
arremessos = df ['Arremessos convertidos'].tolist()

plt.bar(df['Tentativa de arremessos'], df['Arremessos convertidos'], label = 'Relação entre os arremessos convertidos e as tentativas')


plt.xlabel('Tentativa de arremessos')
plt.ylabel('Arremessos convertidos')
plt.xticks([400, 500, 600, 700, 800, 900])
plt.title('Arremessos convertidos')
plt.yticks([400, 500, 600])
plt.legend(loc='upper right')
plt.show()
```

```python
#Tentativas de arremesso e arremessos convertidos
jogadores = ['Stephen Curry', 'Bradley Beal', 'Joel Embiid', 'Damian Lillard', 'Luka Doncic']
tentativa_de_arremesso = [1046, 1086, 711, 1046, 1.068]
arremesso_convertido = [514, 526, 364, 460, 519]

x = np.arange(len(jogadores))  #localizações dos jogadores no grafico
largura = 0.35  #largura das barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - largura/2, tentativa_de_arremesso, largura, label='Tentativas de lançamento')
rects2 = ax.bar(x + largura/2, arremesso_convertido, largura, label='Lançamentos Convertidos')

ax.set_ylabel('Tentativas')
ax.set_title('Lançamentos')
ax.set_xticks(x)
ax.set_xticklabels(jogadores)
ax.legend()

fig.tight_layout()

plt.show()
```
